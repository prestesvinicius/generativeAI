```python
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

genai.configure(api_key='GOOGLE_API_KEY')
```

```python
# Listando os modelos disponíveis
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

```python
generation_config = {
  "candidate_count": 1, # Quantidade de respostas
  "temperature": 0.5, # Quanto mais alto, maior a criatividade da resposta.
}
```

```python
safety_settings={ #Filtros de segurança
    'HATE': 'BLOCK_NONE', # BLOCK_FEW, BLOCK_SOME BLOCK_MOST
    'HARASSMENT': 'BLOCK_NONE',
    'SEXUAL' : 'BLOCK_NONE',
    'DANGEROUS' : 'BLOCK_NONE'
    }
```

```python
# Carregando o modelo escolhido com as configurações selecionadas
model = genai.GenerativeModel(model_name='gemini-1.0-pro',
                                  generation_config=generation_config,
                                  safety_settings=safety_settings,)
```

```python
# Testar o modelo
response = model.generate_content("Me ajude com ideias para um chatbot usando GEMINI API")
print(response.text)
```

```python
response = model.generate_content("Que empresa criou o modelo de IA Gemini?")
print(response)
```

```python
# Historico do modelo, permite lembrar o contexto
chat = model.start_chat(history=[])

# Estrutura do chatbot Q&A
prompt = input('Esperando prompt: ')

while prompt != "fim":
  response = chat.send_message(prompt)
  print("Resposta:", response.text, '\n\n')
  prompt = input('Esperando prompt: ')
```

```python
#Melhorando a visualização
#Código disponível em https://ai.google.dev/tutorials/python_quickstart#import_packages
import textwrap
from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
chat = model.start_chat(history=[])

prompt = input('Esperando prompt: ')

while prompt != "fim":
  response = chat.send_message(prompt)
  display(to_markdown(response.text))
  prompt = input('Esperando prompt: ')
```
